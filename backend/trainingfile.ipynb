```python
import pandas as pd
```

```python
data=pd.read_csv("meradata.csv", delimiter="|")
```

```python
data
```

```python
data.describe()
```

```python
data.rename(columns={"id.orig_h":"originIP", "id.orig_p":"originPort", "id.resp_p":"respPort", "id.resp_h":"respIP"}, inplace=True)
```

```python
data
```

```python
data.columns
```

```python
data_new = data.drop(['uid', 'service', 'resp_pkts','resp_bytes', 'resp_ip_bytes','tunnel_parents','missed_bytes','local_orig', 'local_resp'], axis=1)
```

```python
data_new
```

```python
data_new.replace('-', pd.NA, inplace=True)
```

```python
data_new.isna().sum()
```

```python
data_new["duration"].replace(pd.NA, 0, inplace=True)
```

```python
data_new["orig_bytes"].replace(pd.NA, 0, inplace=True)
```

```python
data_new[(data_new["orig_bytes"]!=0)]
```

```python
data_new["duration"]
```

```python
import matplotlib.pyplot as plt
```

```python
data_final=data_new.drop(["history", "detailed-label"],axis=1)
```

```python
data_final=data_final.drop(["originIP", "respIP"],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
proto_encoder=LabelEncoder()
```

```python
conn_state_encoder=LabelEncoder()
```

```python
data_final["proto"]=proto_encoder.fit_transform(data_final["proto"])
```

```python
data_final["label"]=label_encoder.fit_transform(data_final["label"])
```

```python
data_final["conn_state"]=conn_state_encoder.fit_transform(data_final["conn_state"])
```

```python
data_final.drop(['originPort', 'respPort', 'ts', 'orig_ip_bytes'], axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
data_final
```

```python
train, test= train_test_split(data_final, test_size=0.3)
```

```python
X_train=train.drop(["label"], axis=1)
X_test=test.drop(["label"], axis=1)
y_train=train["label"]
y_test=test["label"]
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
X_train.dtypes
```

```python
model.fit(X=X_train,y=y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(y_test,y_pred)
```

```python
acc
```

```python
from sklearn.metrics import f1_score, precision_recall_curve, roc_auc_score
```

```python
f1=f1_score(y_test,y_pred)
```

```python
f1
```

```python
roc_auc_score(y_test,y_pred)
```
